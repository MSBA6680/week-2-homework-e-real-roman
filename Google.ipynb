```python
import pandas as pd
import numpy as np
import yfinance as yf
import matplotlib.pyplot as plt
```

```python
df = yf.download("GOOG", start="2016-04-01", end="2026-04-01")
df = df[['Close']]
```

```python
df['Returns'] = df['Close'].pct_change()

df['Volatility'] = df['Returns'].rolling(window=252).std() * np.sqrt(252)
```

```python
plt.figure(figsize=(14, 10))

plt.subplot(3, 1, 1)
plt.plot(df.index, df['Close'], label='GOOG Prices')
plt.title('Alphabet (GOOG) Stock Prices')
plt.ylabel('Price')
plt.legend()

plt.subplot(3, 1, 2)
plt.plot(df.index, df['Returns'], label='GOOG Returns')
plt.title('Alphabet Daily Returns')
plt.ylabel('Returns')
plt.legend()

plt.subplot(3, 1, 3)
plt.plot(df.index, df['Volatility'], label='GOOG Annualized Volatility')
plt.title('Alphabet Annualized Volatility')
plt.ylabel('Volatility')
plt.xlabel('Date')
plt.legend()

plt.tight_layout()
plt.show()
```
