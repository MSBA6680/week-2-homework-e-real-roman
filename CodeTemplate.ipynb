```python
%pip install pandas
%pip install numpy
%pip install yfinance
%pip install matplotlib.pyplot
import pandas as pd
import numpy as np
import yfinance as yf
import matplotlib.pyplot as plt
```

```python
df = yf.download("AMZN", start="2016-04-01", end="2026-04-01")
```

```python
df['Returns'] = df['Close'].pct_change()

df['Volatility'] = df['Returns'].rolling(window=252).std() * np.sqrt(252)
```

```python
plt.figure(figsize=(14, 10))

plt.subplot(3, 1, 1)
plt.plot(df.index, df['Close'], label='Amazon Prices')
plt.title('Amazon (AMZN) Stock Prices')
plt.ylabel('Price')
plt.legend()

plt.subplot(3, 1, 2)
plt.plot(df.index, df['Returns'], label='Amazon Returns')
plt.title('Amazon Daily Returns')
plt.ylabel('Returns')
plt.legend()
```

```python
plt.subplot(3, 1, 3)
plt.plot(df.index, df['Volatility'], label='Amazon Annualized Volatility')
plt.title('Amazon Annualized Volatility')
plt.ylabel('Volatility')
plt.xlabel('Date')
plt.legend()

plt.tight_layout()
plt.show()
```
